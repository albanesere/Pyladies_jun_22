Compreender o significado de cada objeto do Pyhton.

Coleo: estrutura que permite armazenar mltiplos itens ou objetos dentro de uma nica unidade.

Tipos de colees:
   - Listas: Estrutura de dados indexados e armazenados em sequncia, onde cada elemento possui uma posio que  identificada por um ndice. O primeiro elemento  esquerda da lista  o ndice 0 e em cada elemento da sequncia o valor do ndice  incrementado (+1). 

      Quando a indexao ocorre da direita para a esquerda, os itens recebem valores negativos, com decremento (-1). 

      Para criar uma coleo do tipo lista no Python: adicionar os itens entre colchetes e separados por vrgula.

      Uma lista pode armazenar qualquer tipo de dado primitivo (string, int, float, etc.). Tambm  possvel criar listas dentro de listas(nested list).[https://docs.python.org/3/tutorial/datastructures.html#more-on-lists]
      Listas devem ser utilizadas ao trabalhar com estruturas homogneas, ou seja, quando todos os elementos so do mesmo tipo (strings, int, float, etc.), e/ou quando existe a necessidade de alterar os itens da coleo. As listas so mutveis, i.., depois de prontas,  possvel substituir valores dentro delas.

   - Tuplas: Estrutura semelhante s listas, com a diferena que so imutveis, i.., depois de prontas no  possve substituir valores dentro delas.


      Para criar uma coleo do tipo tupla no Python: adicionar os itens entre parnteses, e, quando a tupla possuir apenas um item, deve ser adicionada uma vrgula aps o elemento para que o Python reconhea o objeto como tupla, e no como um objeto do tipo String.
[https://docs.python.org/3/tutorial/datastructures.html#tuples-and-sequences]

      Tuplas podem ser utilizadas quando os itens so pr-definidos e no necessitarem de alteraes, ou, quando os dados forem heterogneos (elementos de tipos diferentes).


   - Sets: colees de elementos no-ordenados que no permite itens duplicados (elementos nicos).

      Para declarar um set, os elementos devem estar entre chaves e separados por vrgula. Para criar um set vazio,  necessrio utilizar a funo set().
      
      Os sets so mutveis, assim,  permitido alterar, excluir ou adicionar itens. Contudo, os itens do set devem ser imutveis (string, int ou tuplas).

      Os sets possuem mtodos para operaes matemticas como unio, interseo, diferena e diferena assimtrica. 
https://docs.python.org/3/tutorial/datastructures.html#sets
      
      Sets devem ser utilizados quandofor necessrio trabalhar com elementos nicos, ou ainda, quando existir a necessidade de realizar operaes matemticas com elementos de diferentes sets. Importante notar que, como o set pode ser no-ordenado, os elementos podem estar em ordem diferente a cada execuo do cdigo.


   - [Dicionrios](https://medium.com/ensina-ai/entenda-o-que-e-e-quais-as-diferencas-entre-listas-sets-tuplas-e-dicionarios-no-python-aa927ed64796): estruturas que compreendem um conjunto de pares: chave e valor. Cada chave individual possui um valor associado (essa associao  chamada de coleo associativa desordenada).
      
      Dicionrios so mutveis. 
      O contedo do dicionrio  delimitado por chaves e os elementos so separados por vrgula.
      Para acessar os itens do dicionrio, deve-se utilizar a chave associativa.
https://docs.python.org/3/tutorial/datastructures.html#dictionaries
      Dicionrios devem ser utilizados quando existe a necessidade de armazenar dados de forma organizada (mapear o valor de um elemento com uma chave nica).

   - DataFrame:  uma estrutura de dados bidimensional com os dados alinhados de forma tabular em linhas e colunas, mutvel em tamanho e heterognea.
      
      Os nomes de colunas e os nmeros de linha so conhecidos como ndice de coluna e linha.
      
      As colunas possuem nomes (ndice da coluna) e, as linhas podem ter nomes (ndices textuais) ou podem, por padro, ser numeradas (ndice numrico).
