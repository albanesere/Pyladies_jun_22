Construir listas, questionários e novas variáveis.

Compreender o significado de cada objeto do Pyhton.

Coleção: estrutura que permite armazenar múltiplos itens ou objetos dentro de uma única unidade.

Tipos de coleções:
   - Listas: Estrutura de dados indexados e armazenados em sequência, onde cada elemento possui uma posição que é identificada por um índice. O primeiro elemento à esquerda da lista é o índice 0 e em cada elemento da sequência o valor do índice é incrementado (+1). 

      Quando a indexação ocorre da direita para a esquerda, os itens recebem valores negativos, com decremento (-1). 

      Para criar uma coleção do tipo lista no Python: adicionar os itens entre colchetes e separados por vírgula.

      Uma lista pode armazenar qualquer tipo de dado primitivo (string, int, float, etc.). Também é possível criar listas dentro de listas(nested list).[https://docs.python.org/3/tutorial/datastructures.html#more-on-lists]
      Listas devem ser utilizadas ao trabalhar com estruturas homogêneas, ou seja, quando todos os elementos são do mesmo tipo (strings, int, float, etc.), e/ou quando existe a necessidade de alterar os itens da coleção. As listas são mutáveis, i.é., depois de prontas, é possível substituir valores dentro delas.

   - Tuplas: Estrutura semelhante às listas, com a diferença que são imutáveis, i.é., depois de prontas não é possíve substituir valores dentro delas.


      Para criar uma coleção do tipo tupla no Python: adicionar os itens entre parênteses, e, quando a tupla possuir apenas um item, deve ser adicionada uma vírgula após o elemento para que o Python reconheça o objeto como tupla, e não como um objeto do tipo String.
[https://docs.python.org/3/tutorial/datastructures.html#tuples-and-sequences]

      Tuplas podem ser utilizadas quando os itens são pré-definidos e não necessitarem de alterações, ou, quando os dados forem heterogêneos (elementos de tipos diferentes).


   - Sets: coleções de elementos não-ordenados que não permite itens duplicados (elementos únicos).

      Para declarar um set, os elementos devem estar entre chaves e separados por vírgula. Para criar um set vazio, é necessário utilizar a função set().
      
      Os sets são mutáveis, assim, é permitido alterar, excluir ou adicionar itens. Contudo, os itens do set devem ser imutáveis (string, int ou tuplas).

      Os sets possuem métodos para operações matemáticas como união, interseção, diferença e diferença assimétrica. 
https://docs.python.org/3/tutorial/datastructures.html#sets
      
      Sets devem ser utilizados quandofor necessário trabalhar com elementos únicos, ou ainda, quando existir a necessidade de realizar operações matemáticas com elementos de diferentes sets. Importante notar que, como o set pode ser não-ordenado, os elementos podem estar em ordem diferente a cada execução do código.


   - [Dicionários](https://medium.com/ensina-ai/entenda-o-que-e-e-quais-as-diferencas-entre-listas-sets-tuplas-e-dicionarios-no-python-aa927ed64796): estruturas que compreendem um conjunto de pares: chave e valor. Cada chave individual possui um valor associado (essa associação é chamada de coleção associativa desordenada).
      
      Dicionários são mutáveis. 
      O conteúdo do dicionário é delimitado por chaves e os elementos são separados por vírgula.
      Para acessar os itens do dicionário, deve-se utilizar a chave associativa.
https://docs.python.org/3/tutorial/datastructures.html#dictionaries
      Dicionários devem ser utilizados quando existe a necessidade de armazenar dados de forma organizada (mapear o valor de um elemento com uma chave única).

   - DataFrame: é uma estrutura de dados bidimensional com os dados alinhados de forma tabular em linhas e colunas, mutável em tamanho e heterogênea.
      
      Os nomes de colunas e os números de linha são conhecidos como índice de coluna e linha.
      
      As colunas possuem nomes (índice da coluna) e, as linhas podem ter nomes (índices textuais) ou podem, por padrão, ser numeradas (Índice numérico).
